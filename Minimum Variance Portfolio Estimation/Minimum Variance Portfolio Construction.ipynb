```python
!pip install yfinance
import yfinance as yf
from datetime import datetime
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# Tickers list
# We can add and delete any ticker from the list to get desired ticker live data
ticker =['AAPL', 'AMZN', 'BA', 'BAC', 'BLK', 'LUV', 'COST', 
             'DIS' , 'GOOG', 'GOOS', 'JPM', 'KO', 'LMT', 'MA', 'META' ,
             'MSFT', 'NKE', 'PSX', 'SBUX', 'TSLA']


# We can get data by our choice by giving days bracket
start_date = datetime(2020,1,3)
end_date = datetime(2022,10,23)

data = yf.download(ticker, start=start_date, end=end_date)
df = pd.DataFrame(data) 
df.tail()

```

```python
TSLA = df.iloc[:,19]
TSLA
```

```python
Merged = df.iloc[:,0:20]
Merged
```

```python
TSLA_return = TSLA.pct_change(periods=20)
TSLA_return = TSLA_return.fillna(0)
TSLA_return
```

```python
Merged_return = Merged.pct_change(periods=20)
Merged_return = Merged_return.fillna(0)
Merged_return.tail()
```

```python
Merged_return_diff = Merged_return.sub(Merged_return.iloc[:,19], axis=0)
```

```python
Merged_return_diff = Merged_return_diff.multiply(-1)
```

```python
# Need to split the data into test set and training set
TSLA_return_test = TSLA_return.iloc[-20:,]
TSLA_return_train = TSLA_return.iloc[0:-20,]
Merged_return_test = Merged_return_diff.iloc[-20:,:]
Merged_return_train = Merged_return_diff.iloc[0:-20,:]
```

```python
# Generating Min-Var Portfolio via OLS with Training set
from sklearn.linear_model import LinearRegression

y = np.asarray(TSLA_return_train).reshape(-1,1)
X = np.asarray(Merged_return_train).reshape(-1, 20)
reg = LinearRegression().fit(X, y)
R2 = reg.score(X, y)
beta_aal = reg.coef_
print("OLS Model Coefficents:", beta_aal)
```

```python
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Test out-of-sample performance with the OLS fitted Portfolio
print(reg.predict(Merged_return_test))
```

```python
# Generating Min-Var Portfolio via Ridge Regression with Training set
from sklearn.linear_model import Ridge
rr = Ridge(alpha=5)
rr.fit(X, y)
w = rr.coef_
print("Ridge Regression Model Coefficients:", w)
print("Ridge Regression Model Intercept", rr.intercept_)
```

```python
# Test out-of-sample performance with the Ridge Regression fitted Portfolio
print(rr.predict(Merged_return_test))
```

```python
# Generating Min-Var Portfolio via Lasso Regression with Training set
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
print("Lasso Regression Model Coefficients:", lasso.coef_)
print("Lasso Regression Model Intercept:", lasso.intercept_)
```

```python
# Test out-of-sample performance with the Lasso Regression fitted Portfolio
print(lasso.predict(Merged_return_test))
```

```python
# Generating Min-Var Portfolio via ElasticNet with Training set
from sklearn.linear_model import ElasticNet
regr = ElasticNet(random_state=0)
regr.fit(X, y)
print("Elastic-Net Model Coefficients:", regr.coef_)
print("Elastic-Net Model Intercept:", regr.intercept_)
```

```python
# Test out-of-sample performance with the Elastic Net fitted Portfolio
print(regr.predict(Merged_return_test))
```
